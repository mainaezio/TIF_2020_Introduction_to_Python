```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.01, 4.0, 0.01)

fig, ax = plt.subplots(figsize=(10, 12))

ax.plot(t,t**2,c='b')
#ax.set_ylabel('y = x*x')
#ax.set_xlabel('x')
# sample points
x1 = 1.9
x2 = 2.3
x3 = 2.9
y1 = x1*x1
y2 = x2*x2
y3 = x3*x3
xval = np.array([x1,x2,x3])
yval = np.array([y1,y2,y3])
ax.scatter(xval,yval, marker='o', c='r') 
nullval = np.zeros(3)
ax.scatter(xval,nullval, marker='o', c='r') 
ax.scatter(nullval,yval, marker='o', c='r')
# additional lines
ax.plot([x1,x1],[0.,y1],c='k',linewidth=1.,linestyle='--')
ax.plot([x2,x2],[0.,y2],c='k',linewidth=1.,linestyle='--')
ax.plot([x3,x3],[0.,y3],c='k',linewidth=1.,linestyle='--')
ax.plot([0.,x1],[y1,y1],c='k',linewidth=1.,linestyle='--')
ax.plot([0.,x2],[y2,y2],c='k',linewidth=1.,linestyle='--')
ax.plot([0.,x3],[y3,y3],c='k',linewidth=1.,linestyle='--')

#ax1 = plt.Axes(fig, [0., 0., 1., 1.])
#ax.set_axis_off()
#fig.add_axes(ax1)

## Eliminates all information for x axis
#cur_axes = plt.gca()
#cur_axes.axes.get_xaxis().set_visible(False)

# Eliminates top and right border lines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminates ticks and labels for each axis
ax.axes.get_xaxis().set_visible(False)
ax.axes.get_yaxis().set_visible(False)

#ax.axhline(y=0, color='k')
#ax.axvline(x=0, color='k')

plt.text(x1 - 0.1, -.70, '$r/g$', {'color': 'k', 'fontsize': 20})
plt.text(x2 - 0.03, -.70, '$r$', {'color': 'k', 'fontsize': 20})
plt.text(x3 - 0.03, -.70, '$g$', {'color': 'k', 'fontsize': 20})

plt.text(-.50, y1 - 0.1, '$(r/g)^2$', {'color': 'k', 'fontsize': 20})
plt.text(-.30, y2 - 0.1, 'q', {'color': 'k', 'fontsize': 20})
plt.text(-.30, y3 - 0.1, '$g^2$', {'color': 'k', 'fontsize': 20})

## Saving with small amount of padding around figure
#plt.savefig("Algorithm_My_sqrt_fig.pdf", bbox_inches="tight")  It's complicated to display a pdf in Jupyter
plt.savefig("Algorithm_My_sqrt_fig.png", bbox_inches="tight")
```
